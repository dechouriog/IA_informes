**ALGORITMOS DE BUSQUEDA-Best-First Search**

```python
import heapq #El módulo heapq para implementar colas de prioridad (heaps)
```

```python
class Node: #definición de clase node
    def __init__(self, state, parent=None, action=None, path_cost=0):
        self.state = state #El estado que define el nodo
        self.parent = parent #El nodo padre de donde se origina el nodo actual
        self.action = action #Action tomada desde el padre para llegar al nodo
        self.path_cost = path_cost #costo desde el nodo raiz (estado inicial), hasta el nodo actual

    def __lt__(self, other): #comparar dos objetos de clase node basado en el costo
        return self.path_cost < other.path_cost
```

```python
def expand(problem, node):
    # Completar con la expansión del nodo
    s = node.state
    for action in problem.actions(s):
        s_prime = problem.result(s,action)
        cost = node.path_cost + problem.action_cost(s,action,s_prime)
        yield Node(state = s_prime, parent=node, action = action, path_cost =cost)

    
```

```python
class Problem: #DEFINICION DEL PROBLEMA
    def __init__(self, initial, goal, actions, result, action_cost, is_goal):
        self.initial = initial #Estado inicial
        self.goal = goal #Estado objetivo
        self.actions = actions #acciones disponibles desde un estado.
        self.result = result  #estado resultante de aplicar una acción
        self.action_cost = action_cost #costo de una acción
        self.is_goal = is_goal #verificación de si el estado es el estado objetivo
```

```python
straight_line_distances = {
    'Arad': 366, 'Bucharest': 0, 'Craiova': 160, 'Drobeta': 242,
    'Eforie': 161, 'Fagaras': 176, 'Giurgiu': 77, 'Hirsova': 151,
    'Iasi': 226, 'Lugoj': 244, 'Mehadia': 241, 'Neamt': 234,
    'Oradea': 380, 'Pitesti': 100, 'Rimnicu Vilcea': 193,
    'Sibiu': 253, 'Timisoara': 329, 'Urziceni': 80,
    'Vaslui': 199, 'Zerind': 374
}

def heuristic(state):
    return straight_line_distances.get(state, float('inf'))

def aStar(problem):
    """
    Algoritmo A* para encontrar el camino de costo mínimo desde el estado inicial al objetivo.
    Utiliza la suma del costo del camino y la heurística para priorizar los nodos.
    """
    # Nodo raíz con el estado inicial
    node = Node(state=problem.initial)
    # Frontera como una cola de prioridad: (f(n), nodo)
    frontier = [(node.path_cost + heuristic(node.state), node)]
    heapq.heapify(frontier)
    # Diccionario para registrar los estados alcanzados y su nodo correspondiente
    reached = {problem.initial: node}

    while frontier:
        # Extrae el nodo con el menor valor de f(n)
        _, current = heapq.heappop(frontier)
        # Si es el objetivo, retorna el nodo
        if problem.is_goal(current.state):
            return current
        # Expande el nodo generando sus hijos
        for child in expand(problem, current):
            s = child.state
            # Si el estado no ha sido alcanzado o se alcanza con menor costo
            if s not in reached or child.path_cost < reached[s].path_cost:
                reached[s] = child
                # Calcula f(n) = g(n) + h(n)
                f = child.path_cost + heuristic(child.state)
                heapq.heappush(frontier, (f, child))
    # Si no se encuentra solución
    return None
```
